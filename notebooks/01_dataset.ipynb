```python
from modules.dataset import get_dataloader
import matplotlib.pyplot as plt
```

```python
loader = get_dataloader(batch_size=4)
(x1, x2) = next(iter(loader))

fig, axs = plt.subplots(2, 4, figsize=(8,4))
for i in range(4):
    axs[0, i].imshow(x1[i].permute(1,2,0).numpy() * 0.5 + 0.5)  # denormalize
    axs[0, i].set_title("View 1")
    axs[0, i].axis("off")

    axs[1, i].imshow(x2[i].permute(1,2,0).numpy() * 0.5 + 0.5)
    axs[1, i].set_title("View 2")
    axs[1, i].axis("off")
plt.show()
```
